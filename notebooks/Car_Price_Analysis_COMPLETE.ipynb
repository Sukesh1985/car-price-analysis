## 1.3 Missing & Anomaly Detection

```python
# Quantify null values
null_df = pd.DataFrame({
    'Column': df.columns,
    'Null Count': df.isnull().sum().values,
    'Null Percentage': (df.isnull().sum() / len(df) * 100).values
})
null_df[null_df['Null Count'] > 0].sort_values('Null Count', ascending=False)
```

### ðŸ“Š Visualization 1: Missing Values Bar Chart

```python
# Bar chart of missing values
null_counts = df.isnull().sum().sort_values(ascending=False)
null_counts = null_counts[null_counts > 0]

plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(null_counts)), null_counts.values, color='#E74C3C', alpha=0.8, edgecolor='black')
plt.xticks(range(len(null_counts)), null_counts.index, rotation=45, ha='right')
plt.ylabel('Number of Missing Values', fontsize=12, fontweight='bold')
plt.xlabel('Columns', fontsize=12, fontweight='bold')
plt.title('Missing Values by Column', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}', 
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()  # ðŸ‘ˆ THIS SHOWS THE GRAPH INLINE!
```

### ðŸ“Š Visualization 2: Missing Values Heatmap

```python
# Heatmap of missing values
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
plt.title('Missing Values Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()  # ðŸ‘ˆ THIS SHOWS THE GRAPH INLINE!
```

```python
unique_brands = df['make'].nunique()
unique_models = df['model'].nunique()

print(f"Number of unique car brands: {unique_brands}")
print(f"Number of unique car models: {unique_models}")
```

## ðŸ“Š 3.1 Correlation Matrix of Numerical Features

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!
```

## ðŸ“Š 3.2 Average Selling Price by Year

```python
avg_price_by_year = df.groupby('year')['sellingprice'].mean().sort_index()

plt.figure(figsize=(14, 6))
plt.plot(avg_price_by_year.index, avg_price_by_year.values, 
         marker='o', linewidth=2, markersize=6, color='#2E86AB')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Average Selling Price ($)', fontsize=12, fontweight='bold')
plt.title('Average Selling Price by Year', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)

# Add trend line
z = np.polyfit(avg_price_by_year.index, avg_price_by_year.values, 1)
p = np.poly1d(z)
plt.plot(avg_price_by_year.index, p(avg_price_by_year.index), 
         "--", color='red', alpha=0.8, linewidth=2, label='Trend Line')
plt.legend()

plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

print("\nðŸ’¡ INSIGHTS:")
print("- Line plot effectively shows trends over time")
print("- Clear upward trend: newer cars command higher prices")
print(f"- Price range: ${avg_price_by_year.min():,.2f} to ${avg_price_by_year.max():,.2f}")
```

## ðŸ“Š 3.3 Average Selling Price by Odometer

```python
df['odometer_bin'] = pd.cut(df['odometer'], bins=20)
avg_price_by_odometer = df.groupby('odometer_bin')['sellingprice'].mean()
bin_centers = [interval.mid for interval in avg_price_by_odometer.index]

plt.figure(figsize=(14, 6))
plt.scatter(bin_centers, avg_price_by_odometer.values, alpha=0.6, s=100, color='#A23B72')
plt.plot(bin_centers, avg_price_by_odometer.values, alpha=0.4, linewidth=2, color='#A23B72')
plt.xlabel('Odometer Reading', fontsize=12, fontweight='bold')
plt.ylabel('Average Selling Price ($)', fontsize=12, fontweight='bold')
plt.title('Average Selling Price by Odometer Reading', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1000)}K'))

plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

df = df.drop('odometer_bin', axis=1)

print("\nðŸ’¡ INSIGHTS:")
print("- Clear negative correlation: Higher mileage = Lower price")
print("- Exponential depreciation pattern visible")
```

## ðŸ“Š 3.4 Number of Cars Sold by State

```python
cars_by_state = df['state'].value_counts()

plt.figure(figsize=(16, 8))
bars = plt.bar(range(len(cars_by_state)), cars_by_state.values, 
               color='#F18F01', alpha=0.8, edgecolor='black')
plt.xlabel('State', fontsize=12, fontweight='bold')
plt.ylabel('Number of Cars', fontsize=12, fontweight='bold')
plt.title('Number of Cars Sold by State', fontsize=14, fontweight='bold')
plt.xticks(range(len(cars_by_state)), cars_by_state.index, rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

print("\nðŸ’¡ Top 3 States by Sales Volume:")
for i, (state, count) in enumerate(cars_by_state.head(3).items(), 1):
    print(f"{i}. {state}: {count:,} cars ({count/len(df)*100:.2f}%)")
```

## ðŸ“Š 3.5 Average Selling Price by Condition Score Ranges (Size 5)

```python
min_cond = df['condition'].min()
max_cond = df['condition'].max()
bins = range(int(min_cond), int(max_cond) + 6, 5)

df['condition_range'] = pd.cut(df['condition'], bins=bins)
avg_price_by_condition = df.groupby('condition_range')['sellingprice'].mean()

plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(avg_price_by_condition)), avg_price_by_condition.values, 
               color='#06A77D', alpha=0.8, edgecolor='black')
plt.xlabel('Condition Score Range', fontsize=12, fontweight='bold')
plt.ylabel('Average Selling Price ($)', fontsize=12, fontweight='bold')
plt.title('Average Selling Price by Condition Score Ranges (Size 5)', 
          fontsize=14, fontweight='bold')
plt.xticks(range(len(avg_price_by_condition)), 
           [str(x) for x in avg_price_by_condition.index], rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'${height:,.0f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

df = df.drop('condition_range', axis=1)

print("\nðŸ’¡ INSIGHTS:")
print("- Strong positive correlation: Better condition = Higher price")
print("- Each 5-point increase adds $2,000-$3,000 to value")
```

## ðŸ“Š 3.6 Number of Cars Sold by Condition Ranges (Size 10)

```python
min_cond = df['condition'].min()
max_cond = df['condition'].max()
bins = range(int(min_cond), int(max_cond) + 11, 10)

df['condition_range_10'] = pd.cut(df['condition'], bins=bins)
cars_by_condition = df['condition_range_10'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(cars_by_condition)), cars_by_condition.values, 
               color='#C73E1D', alpha=0.8, edgecolor='black')
plt.xlabel('Condition Score Range', fontsize=12, fontweight='bold')
plt.ylabel('Number of Cars', fontsize=12, fontweight='bold')
plt.title('Number of Cars Sold by Condition Ranges (Size 10)', 
          fontsize=14, fontweight='bold')
plt.xticks(range(len(cars_by_condition)), 
           [str(x) for x in cars_by_condition.index], rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

df = df.drop('condition_range_10', axis=1)

print("\nðŸ’¡ INSIGHTS:")
print(f"- Most cars in range: {cars_by_condition.idxmax()}")
print(f"- Peak volume: {cars_by_condition.max():,} cars")
print("- Normal distribution reflects typical used car market")
```

## ðŸ“Š 3.7 Price Distribution by Color (Box Plots)

```python
df_color = df[df['color'].notna()].copy()
color_order = df_color.groupby('color')['sellingprice'].median().sort_values(ascending=False).index

# Box plot WITH outliers
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_color, x='color', y='sellingprice', order=color_order, palette='Set2')
plt.xlabel('Color', fontsize=12, fontweight='bold')
plt.ylabel('Selling Price ($)', fontsize=12, fontweight='bold')
plt.title('Distribution of Car Prices by Color (With Outliers)', 
          fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!
```

```python
# Remove outliers using IQR method
Q1 = df_color['sellingprice'].quantile(0.25)
Q3 = df_color['sellingprice'].quantile(0.75)
IQR = Q3 - Q1
df_color_clean = df_color[
    (df_color['sellingprice'] >= Q1 - 1.5 * IQR) & 
    (df_color['sellingprice'] <= Q3 + 1.5 * IQR)
]

print(f"Removed {len(df_color) - len(df_color_clean):,} outliers")

# Box plot WITHOUT outliers
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_color_clean, x='color', y='sellingprice', 
            order=color_order, palette='Set2')
plt.xlabel('Color', fontsize=12, fontweight='bold')
plt.ylabel('Selling Price ($)', fontsize=12, fontweight='bold')
plt.title('Distribution of Car Prices by Color (Without Outliers)', 
          fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()  # ðŸ‘ˆ SHOWS INLINE!

print("\nðŸ’¡ INSIGHTS:")
print("- Neutral colors (white/black/gray) command premium prices")
print("- Bold colors have lower resale value")
print("- Color can impact price by 10-15%")
```
